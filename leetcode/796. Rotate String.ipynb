```java
import java.util.ArrayDeque;
import java.util.Arrays;

class Solution <T> {
    public boolean rotateString(String A, String B) {
        if (A.equals("") && B.equals("")) {
            return true;
        }

        ArrayDeque<T> dequeA = (ArrayDeque<T>) stringToDeque(A);
        ArrayDeque<T> dequeB = (ArrayDeque<T>) stringToDeque(B);

        for (int i=0; i < dequeA.size(); ++i) {
            dequeA = this.rotateDeque(dequeA);
            String[] listA;
            String[] listB;
            listA = dequeA.toArray(new String[0]);
            listB = dequeB.toArray(new String[0]);
            if (Arrays.equals(listA, listB)) {
                return true;
            } 
        }

        return false;
    }

    private ArrayDeque<String> stringToDeque(String s) {
        ArrayDeque<String> deque = new ArrayDeque<String>();
        for (int i=0; i<s.length(); ++i) {
            deque.add(s.substring(i, i+1));
        }
        return deque;
    }

    private ArrayDeque<T> rotateDeque(ArrayDeque<T> deque) {
        T left = deque.removeFirst();
        deque.add(left);
        return  deque;
    }

    public static void main(String[] args) {
        Boolean r = new Solution().rotateString("", "");
        System.out.println(r.toString());
    }
}

```
