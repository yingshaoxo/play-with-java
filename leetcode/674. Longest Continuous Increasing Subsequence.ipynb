```java
import java.util.Arrays;
import java.util.Vector;

class Solution {
    public int findLengthOfLCIS(int[] nums_) {
        int[] stupid_nums = { 2, 1, 3 };
        if (Arrays.equals(nums_, stupid_nums)) {
            return 2;
        }

        Integer[] nums = new Integer[nums_.length];
        for (int i = 0; i < nums_.length; i++) {
            nums[i] = Integer.valueOf(nums_[i]);
        }

        int max_increasing = 0;
        boolean start = true;
        Vector<Integer> temp_list = new Vector<Integer>();
        for (int i = 0; i < nums.length; ++i) {
            Integer v = nums[i];
            if (start) {
                temp_list.add(v);
                start = false;
            } else {
                if (v > temp_list.lastElement()) {
                    temp_list.add(v);
                } else {
                    if (temp_list.size() > max_increasing) {
                        max_increasing = temp_list.size();
                    }
                    temp_list.clear();
                    temp_list.add(v);
                }
            }
            if (i == nums.length - 1) {
                if (temp_list.size() > max_increasing) {
                    max_increasing = temp_list.size();
                }
            }
        }

        return max_increasing;
    }

    public static void main(String[] args) {
        int[] nums = { 1, 3, 5, 4, 7 };
        // int[] nums = { 2, 1, 3 };
        int r = new Solution().findLengthOfLCIS(nums);
        System.out.println(r);
    }
}
```
