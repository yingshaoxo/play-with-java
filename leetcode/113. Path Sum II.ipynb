```java
import java.util.List;
import java.util.ArrayList;

class TreeNode {
    int val;
    TreeNode left;
    TreeNode right;

    TreeNode() {
    }

    TreeNode(int val) {
        this.val = val;
    }

    TreeNode(int val, TreeNode left, TreeNode right) {
        this.val = val;
        this.left = left;
        this.right = right;
    }
}

class Solution {
    List<List<Integer>> paths;
    int targtet;

    public List<List<Integer>> pathSum(TreeNode root, int sum) {
        this.paths = new ArrayList<List<Integer>>();
        this.targtet = sum;

        this.travel(root, new ArrayList<Integer>());

        return this.paths;
    }

    public void travel(TreeNode node, List<Integer> l) {
        if (node == null) {
            return; 
        } else {
            l.add(node.val);
            if ((node.left == null) && (node.right == null)) {
                if (this.sum(l) == this.targtet) {
                    List<Integer> copyOfL = new ArrayList<>();
                    copyOfL.addAll(l);
                    this.paths.add(copyOfL);
                }
            }
        }
        int length = l.size();
        this.travel(node.left, l);
        l = l.subList(0, length);
        this.travel(node.right, l);
    }

    private int sum(List<Integer> l) {
        int v = 0;
        for (int i = 0; i < l.size(); i++) {
            v += l.get(i);
        }
        return v;
    } 

    public static void main(String[] args) {
        TreeNode root = new TreeNode(0);
        root.left = new TreeNode(1);
        root.left.left = new TreeNode(1);
        root.left.right = new TreeNode(0);
        root.right = new TreeNode(2);
        root.right.left = new TreeNode(-1);
        root.right.right = new TreeNode(0);
        List<List<Integer>> r = new Solution().pathSum(root, 1);
        System.out.println(r.toString());
    }
}
```
